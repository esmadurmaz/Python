```python
listt_1 = []
inp1 = input('Enter a word > ')
while True:
    inp2 = input("Enter a word for the word list or 'q' to quit > ")
    if inp2.lower() == 'q':
        break
    else:
        listt_1.append(inp2)
listt_2 = []
sonuc = True
for i in listt_1:
    if set(i) == set(inp1):
        for j in i:
            if i.count(j) != inp1.count(j):
                sonuc = False
        if sonuc:
            if i not in listt_2:
                listt_2.append(i)
print(listt_2)
```
